```python
import matplotlib.pyplot as plt

def bipolar_return_zero_encoding(data, bit_duration, voltage_level=1, zero_voltage=0):
    encoded_signal = []
    
    last_polarity = 1  # Initialize the polarity to positive
    
    for bit in data:
        if bit == 0:
            signal = [zero_voltage] * bit_duration
        else:
            signal = [voltage_level * last_polarity] * (bit_duration // 2) + [zero_voltage] * (bit_duration // 2)
            last_polarity *= -1  # Change the polarity for the next bit
        
        encoded_signal.extend(signal)
    
    return encoded_signal

# Example data: binary sequence
data_sequence = [1, 1, 1, 0, 1, 1, 0, 0]

# Parameters
bit_duration = 10  # Duration of each bit in time units (arbitrary)
voltage_level = 1  # Voltage level for '1'
zero_voltage = 0   # Voltage level for '0'

# Generate the encoded signal
encoded_signal = bipolar_return_zero_encoding(data_sequence, bit_duration, voltage_level, zero_voltage)

# Plot the encoded signal
plt.figure(figsize=(10, 3))
plt.plot(encoded_signal)
plt.title("Bipolar Return-to-Zero Encoding")
plt.xlabel("Time")
plt.ylabel("Voltage Level")
plt.ylim(-1.5, 1.5)
#plt.xticks([])  # Hide x-axis ticks for simplicity
#plt.yticks([-1, 0, 1], ['-1', '0', '1'])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
