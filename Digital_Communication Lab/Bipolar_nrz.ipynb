```python
import matplotlib.pyplot as plt

def bipolar_nrz_encoding(data, bit_duration, voltage_level=1, zero_voltage=-1):
    encoded_signal = []
    last_polarity = 1 
    for bit in data:
        if bit == 0:
            signal = [zero_voltage] * bit_duration
        else:
            signal = [voltage_level * last_polarity  ] * bit_duration
            last_polarity *= -1
        
        encoded_signal.extend(signal)
    
    return encoded_signal

# Example data: binary sequence
data_sequence = [1, 1, 1, 0, 1, 1, 0, 0]

# Parameters
bit_duration = 10  # Duration of each bit in time units (arbitrary)
voltage_level = 1  # Voltage level for '1'
zero_voltage = -1  # Voltage level for '0'

# Generate the encoded signal
encoded_signal = bipolar_nrz_encoding(data_sequence, bit_duration, voltage_level, zero_voltage)

# Create time axis for plotting
time_axis = list(range(0, len(encoded_signal) * bit_duration, bit_duration))

# Plot the encoded signal
plt.figure(figsize=(10, 3))
plt.step(time_axis, encoded_signal, where='post', color='blue')
plt.title("Bipolar Non-Return-to-Zero Encoding")
plt.xlabel("Time")
plt.ylabel("Voltage Level")
plt.ylim(-1.5, 1.5)
plt.yticks([-1, 0, 1], ['-1', '0', '1'])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
