```python
import matplotlib.pyplot as plt

def manchester_encode(bits):
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([-1, 1])  # Encode 0 as a high-to-low transition
        else:
            signal.extend([1, -1])  # Encode 1 as a low-to-high transition
    return signal

def plot_signal(signal):
    time = range(len(signal))
    plt.plot(time, signal)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Manchester Signaling')
    plt.ylim([-1.5, 1.5])  # Adjust the y-axis limits if needed
    plt.show()

# Example usage
bits = [1, 0, 1, 1, 0, 1, 0, 0]  # Replace with your own bit sequence
signal = manchester_encode(bits)
plot_signal(signal)
```
