```python
import matplotlib.pyplot as plt
import numpy as np

def create_waveform(bits, duration=1, sampling_rate=44100, frequency=440):
    # Calculate the time array
    t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

    # Create an empty waveform array
    waveform = np.zeros_like(t)

    # Define the high and low amplitudes
    high_amplitude = 1
    low_amplitude = -1

    # Iterate over the bits and set the waveform values
    for i, bit in enumerate(bits):
        if bit == 1:
            waveform[i * sampling_rate:(i + 1) * sampling_rate] = high_amplitude
        else:
            waveform[i * sampling_rate:(i + 1) * sampling_rate] = low_amplitude

    # Generate the carrier wave
    carrier_wave = np.sin(2 * np.pi * frequency * t)

    # Modulate the carrier wave with the binary waveform
    modulated_wave = carrier_wave * waveform

    return t, modulated_wave

# Example usage
bits = [1, 0, 1, 0, 1, 1, 1, 1, 1, 0]  # Sample bit sequence
duration = 1  # Duration of the waveform in seconds
sampling_rate = 44100  # Number of samples per second
frequency = 440  # Frequency of the carrier wave

t, waveform = create_waveform(bits, duration, sampling_rate, frequency)

# Plot the waveform
plt.plot(t, waveform)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Binary Waveform')
plt.show()
```
