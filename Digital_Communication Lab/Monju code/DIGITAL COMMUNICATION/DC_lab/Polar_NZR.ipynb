```python
import matplotlib.pyplot as plt

def polar_nrz_encode(bits):
    signal = []
    for bit in bits:
        if bit == 0:
            signal.extend([-1]*10)  # Insert 10 negative (-1) values for each 0 bit
        else:
            signal.extend([1]*10)  # Insert 10 positive (+1) values for each 1 bit
    return signal

def plot_signal(signal):
    time = range(len(signal))
    plt.plot(time, signal, drawstyle='steps-pre')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Polar NRZ Signaling')
    plt.ylim([-1.5, 1.5])  # Adjust the y-axis limits if needed
    plt.show()

# Example usage
bits = [1, 0, 1, 1, 0, 1, 0, 0]  # Replace with your own bit sequence
signal = polar_nrz_encode(bits)
plot_signal(signal)
```
