```python
import matplotlib.pyplot as plt

def polar_rz_encoding(data, bit_duration, voltage_level=1, zero_voltage=-1):
    encoded_signal = []
    
    for bit in data:
        if bit == 0:
            signal = [zero_voltage] * bit_duration
        else:
            signal = [voltage_level] * (bit_duration // 2) + [zero_voltage] * (bit_duration // 2)
        
        encoded_signal.extend(signal)
    
    return encoded_signal

# Example data: binary sequence
data_sequence = [1, 0, 1, 0, 1, 1, 0, 0]

# Parameters
bit_duration = 10  # Duration of each bit in time units (arbitrary)
voltage_level = 1  # Voltage level for '1'
zero_voltage = -1  # Voltage level for '0'

# Generate the encoded signal
encoded_signal = polar_rz_encoding(data_sequence, bit_duration, voltage_level, zero_voltage)

# Plot the encoded signal
plt.figure(figsize=(10, 3))
plt.plot(encoded_signal)
plt.title("Polar RZ Encoding")
plt.xlabel("Time")
plt.ylabel("Voltage Level")
plt.ylim(-1.5, 1.5)
#plt.xticks([])  # Hide x-axis ticks for simplicity
#plt.yticks([-1, 0, 1], ['-1', '0', '1'])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
