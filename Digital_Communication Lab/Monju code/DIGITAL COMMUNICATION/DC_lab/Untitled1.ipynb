```python
import numpy as np
import matplotlib.pyplot as plt

def ask_modulation(data_bits, carrier_frequency, bit_rate, amplitude=1.0, sampling_rate=1000):
    t_bit = 1 / bit_rate
    t = np.arange(0, len(data_bits) * t_bit, 1 / sampling_rate)
    
    carrier_signal = amplitude * np.sin(2 * np.pi * carrier_frequency * t)

    modulated_signal = []
    for bit in data_bits:
        if bit == 0:
            modulated_signal.extend([0] * int(t_bit * sampling_rate))
        else:
            modulated_signal.extend(carrier_signal[:int(t_bit * sampling_rate)])
    
    return t, np.array(modulated_signal)

if __name__ == "__main__":
    data = input("Enter binary data (e.g., 010101): ")
    data_bits = [int(bit) for bit in data]

    carrier_frequency = float(input("Enter carrier frequency (Hz): "))
    bit_rate = float(input("Enter bit rate (bits per second): "))

    t, modulated_signal = ask_modulation(data_bits, carrier_frequency, bit_rate)

    plt.plot(t, modulated_signal)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.title('Amplitude Shift Keying (ASK) Modulation')
    plt.grid(True)
    plt.show()
```
