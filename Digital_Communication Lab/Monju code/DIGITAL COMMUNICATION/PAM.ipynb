```python
import numpy as np
import matplotlib.pyplot as plt

fc = 100
fm = fc / 10
fs = 100 * fc
t = np.arange(0, 4 / fm, 1 / fs)
mt = np.cos(2 * np.pi * fm * t)
ct = 0.5 * (np.sign(np.sin(2 * np.pi * fc * t)) + 1)  # Equivalent to square wave in MATLAB
st = mt * ct
tt = []

# Single-sided PAM
for i in range(len(st)):
    if st[i] == 0:
        tt.append(st[i])
    else:
        tt.append(st[i] + 2)

tt = np.array(tt)

plt.figure(figsize=(10, 12))

plt.subplot(4, 1, 1)
plt.plot(t, mt)
plt.title('Message Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(t, ct)
plt.title('Carrier Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 3)
plt.plot(t, st)
plt.title('Modulated Signal of Double Side Band')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 4)
plt.plot(t, tt)
plt.title('PAM of Single Side Band')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
