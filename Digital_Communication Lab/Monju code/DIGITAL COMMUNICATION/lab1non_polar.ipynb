```python
import matplotlib.pyplot as plt

def unipolar_nrz_encode(binary_sequence):
    signal = []
    for bit in binary_sequence:
        signal.extend([1 if bit == '1' else 0] * 10)  # Extend each bit for 10 time units
    return signal

def plot_signal(signal, title):
    plt.step(range(0, len(signal)), signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Signal')
    plt.ylim([-0.5, 1.5])
    plt.show()

def main():
    binary_sequence = input("Enter binary sequence (e.g., 101010): ")
    encoded_signal = unipolar_nrz_encode(binary_sequence)
    
    print("Encoded Signal:", encoded_signal)
    
    plot_signal(encoded_signal, 'Unipolar NRZ Signaling')

if __name__ == "__main__":
    main()
```
