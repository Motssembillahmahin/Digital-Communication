```python
import matplotlib.pyplot as plt

def polar_nrz_encode(binary_sequence):
    signal = []
    for bit in binary_sequence:
        if bit == '0':
            signal.extend([-1] * 10)  # Extend each '0' bit with a negative voltage level for 10 time units
        else:
            signal.extend([1] * 10)   # Extend each '1' bit with a positive voltage level for 10 time units
    return signal

def plot_signal(signal, title):
    plt.step(range(0, len(signal)), signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Signal')
    plt.ylim([-1.5, 1.5])
    plt.show()

def main():
    binary_sequence = input("Enter binary sequence (e.g., 101010): ")
    encoded_signal = polar_nrz_encode(binary_sequence)
    
    print("Encoded Signal:", encoded_signal)
    
    plot_signal(encoded_signal, 'Polar NRZ Signaling')

if __name__ == "__main__":
    main()
```
