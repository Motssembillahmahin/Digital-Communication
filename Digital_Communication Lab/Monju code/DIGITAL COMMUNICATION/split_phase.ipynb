```python
import matplotlib.pyplot as plt

def split_phase_encode(binary_sequence):
    signal = []
    for bit in binary_sequence:
        if bit == '1':
            signal.extend([-1, 1] * 5)  # Transition from negative to positive for '1' bit
        else:
            signal.extend([1, -1] * 5)  # Transition from positive to negative for '0' bit
    return signal

def plot_signal(signal, title):
    plt.step(range(0, len(signal)), signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Signal')
    plt.ylim([-1.5, 1.5])
    plt.show()

def main():
    binary_sequence = input("Enter binary sequence (e.g., 101010): ")
    encoded_signal = split_phase_encode(binary_sequence)
    
    print("Encoded Signal:", encoded_signal)
    
    plot_signal(encoded_signal, 'Split-Phase (Manchester Coding)')

if __name__ == "__main__":
    main()
```
