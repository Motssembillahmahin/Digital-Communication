```python
import matplotlib.pyplot as plt

def unipolar_rz_encode(binary_sequence):
    signal = []
    for bit in binary_sequence:
        if bit == '1':
            signal.extend([1] * 10)  # Extend each '1' bit with a positive voltage level for 10 time units
            signal.extend([0] * 10)  # Insert a zero voltage level for 10 time units (signal return to zero)
        else:
            signal.extend([0] * 20)  # Extend each '0' bit with a zero voltage level for 20 time units
    return signal

def plot_signal(signal, title):
    plt.step(range(0, len(signal)), signal, where='post')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Signal')
    plt.ylim([-0.5, 1.5])
    plt.show()

def main():
    binary_sequence = input("Enter binary sequence (e.g., 101010): ")
    encoded_signal = unipolar_rz_encode(binary_sequence)
    
    print("Encoded Signal:", encoded_signal)
    
    plot_signal(encoded_signal, 'Unipolar RZ Line Coding')

if __name__ == "__main__":
    main()
```
