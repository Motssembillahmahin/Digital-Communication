```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([1, 0, 1, 0, 1, 1, 0, 0, 1, 0, 1, 1, 0])

last = len(data)
t = 0
q = np.array([])
o = np.array([])

for start in range(last):
    if data[start] == 0:  # 00
        for j in np.arange(0, 20, 0.1):
            q = np.append(q, 0)
            o = np.append(o, np.sin(j))  # 0 degree
            
    elif data[start] == 1:  # 01
        for j in np.arange(0, 20, 0.1):
            q = np.append(q, 1)
            o = np.append(o, np.cos(j))  # 90 degree
            
    elif data[start] == 10:  # 10
        for j in np.arange(0, 20, 0.1):
            q = np.append(q, 2)
            o = np.append(o, -np.sin(j))  # 180 degree
            
    else:  # data[start] == 11 (11)
        for j in np.arange(0, 20, 0.1):
            q = np.append(q, 3)
            o = np.append(o, -np.cos(j))  # 270 degree

plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(q)
plt.title('INPUT')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xlim(0, 2600)
plt.ylim(-3, 3)
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(o)
plt.title('OUTPUT')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid()

plt.tight_layout()
plt.show()
```
