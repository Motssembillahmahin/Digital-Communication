```python
import matplotlib.pyplot as plt

def unipolar_nrz_encode(bits):
    signal = []
    for bit in bits:
        signal.extend([bit]*10)  # Repeat each bit 10 times (adjust the factor as needed)
    return signal

def plot_signal(signal):
    time = range(len(signal))
    plt.plot(time, signal, drawstyle='steps-pre')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title('Unipolar NRZ Signaling')
    plt.ylim([-0.5, 1.5])  # Adjust the y-axis limits if needed
    plt.show()

# Example usage
bits = [1, 0, 1, 1, 0, 1, 0, 0]  # Replace with your own bit sequence
signal = unipolar_nrz_encode(bits)
plot_signal(signal)
```
