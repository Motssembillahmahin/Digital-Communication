```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

def binary():
    sample_size = 32
    binary_code = [1,0,1,0,1,0,1,0,1]
    binary_code_arr = np.array(binary_code)
    b_len = len(binary_code_arr)
    #plt.stem(binary_code_arr)

    #generating signal
    signal = np.zeros(sample_size*b_len)

    id_n = np.where(binary_code_arr == 1)

    for i in id_n[0]:
        temp = int(i*sample_size)
        signal[temp:temp+sample_size] = 1
    
    return signal

sig = binary()

sampling_rate = 1000  # Sampling rate in Hz
carrier_freq = 10    # Frequency of the carrier signal in Hz
bit_rate = 2   
t = np.arange(0, 1, 1/sampling_rate)
carrier_signal = 10*np.sin(2*pi*carrier_freq*t)


plt.plot(sig)
plt.plot(carrier_signal)
```
