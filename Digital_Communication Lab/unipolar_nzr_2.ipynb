```python
import matplotlib.pyplot as plt
import numpy as np

binary_code = [0, 1, 0, 1, 0, 1, 0, 0]
access_bit = 1
t = np.arange(0,len(binary_code),0.01)
p = [] ##push the amplitude value
for i in range(0, len(binary_code)*100, 1) :
    if t[i] < access_bit :
        p.append(binary_code[access_bit - 1])
    else :
        p.append(binary_code[access_bit])
        access_bit = access_bit + 1 
plt.plot(t,p)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Unipolar NRZ')
plt.ylim([-0.5, 1.5])
```
