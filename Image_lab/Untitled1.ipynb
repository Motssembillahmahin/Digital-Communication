```python
import matplotlib.pyplot as plt
import cv2 
import numpy as np
```

```python
img = cv2.imread('image_2.jpg')
#cv2.imshow(img)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
h = np.zeros((256), dtype = int)
k = 0
while k < 256:
    h[k] = np.count_nonzero(gray == k)
    k = k + 1
intensity = np.arange(0, 256, 1)
plt.bar(intensity, h, color = 'blue', width = .5)
    
```
